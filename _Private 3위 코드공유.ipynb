## 결측치 개수 확인

결측치를 확인하는 여러 방법이 있습니다. `info()` 메소드의 Non-Null Count 부분 `describe()` 메소드의 count 부분으로 간단히 결측치의 존재 유무를 확인할 수 있었죠? 이번에는 결측치의 개수를 확인할 수 있는 메소드를 알아봅시다.

`isnull()` 메소드는 결측치(NaN)면 True, 그렇지 않으면 False를 반환합니다. 

컴퓨터는 True를 1로, False는 0으로 인식합니다. 이를 활용하여 `sum()`을 적용하면 결측치의 개수를 알 수 있습니다.

```python
# 결측치 개수 확인
train.isnull().sum()
```

이번 데이터에는 결측치가 존재하지 않으므로 모두 0이 나옵니다!

**1) 법선 + 수평 일사량을 합친다면 발전량과 비례할 것이다.**

```python
# DNI + DHI 합친 컬럼 SUN 을 만들어준다.
train = ( train
 .assign( SUN = lambda df: train['DHI'] + train['DNI'] )
 )

test = (test
    .assign( SUN = lambda df: test['DHI'] + test['DNI'] )
 )

sns.lmplot(data = train, x='SUN', y='TARGET')
plt.show()
```

일사량을 합친 SUN 칼럼은 발전량과 높은 연관성을 보인다.

**2) 일사량과 온도를 합친 컬럼은 발전량과 비례할 것이다.**

```python
# 일사량 + 온도를 합친 칼럼 SS 추가
train = (train
 .assign( SS = lambda df: train['SUN'] + train['T'] )
 )

test = (test
    .assign( SS = lambda df: test['SUN'] + test['T'] )
 )
sns.lmplot(data = train, x='SS', y='TARGET')
plt.show()
```

 **3)법선 일사량과 상대습도는 반비례할 것이다.**

```python
# 법선 일사량과 상대습도를 합친 칼럼 DRH 추가
train = (train
 .assign( DRH = lambda df: train['DNI'] +  ( -1 * train['RH'] ) )
 )

test = (test
 .assign( DRH = lambda df: test['DNI'] +  ( -1 * test['RH'] ) )
 )

sns.lmplot(data = train, x='TARGET', y='DRH')
plt.show()
```

법선 일사량과 상대습도는 비례한다.
